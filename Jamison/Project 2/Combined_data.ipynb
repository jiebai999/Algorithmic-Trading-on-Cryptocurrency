```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARIMAResults

%matplotlib inline
```

```python
btc_model1 = ARIMAResults.load('btc_model1.pkl')
```

```python
btc_model1.summary()
```

```python
eth_model1 = ARIMAResults.load('eth_model1.pkl')
```

```python
doge_model1 = ARIMAResults.load('doge_model1.pkl')
```

```python
ltc_model1 = ARIMAResults.load('ltc_model1.pkl')
```

```python
btc_plot = pd.DataFrame(btc_model1.forecast(steps=30)[0]).plot(title="BTC Futures Forecast 1")
eth_plot = pd.DataFrame(eth_model1.forecast(steps=30)[0]).plot(title="ETH Futures Forecast 1")
doge_plot = pd.DataFrame(doge_model1.forecast(steps=30)[0]).plot(title="DOGE Futures Forecast 1")
ltc_plot = pd.DataFrame(ltc_model1.forecast(steps=30)[0]).plot(title="LTC Futures Forecast 1")
```

```python
btc_plot.figure.savefig('BTC_Futures1.png')
eth_plot.figure.savefig('ETH_Futures1.png')
doge_plot.figure.savefig('DOGE_Futures1.png')
ltc_plot.figure.savefig('LTC_Futures1.png')
```
